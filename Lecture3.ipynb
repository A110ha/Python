```python
def word_frequency(text):
    text = text.lower()
    punctuation = '.,!?;:"()[]{}<>/\\@#$%^&*+-_=|'
    for char in punctuation:
        text = text.replace(char, ' ')
    words = text.split()
    frequency_dict = {}
    for word in words:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    return frequency_dict
text = "Привет, как дела? Как дела, привет!"
result = word_frequency(text)
print(result)
```

```python
def merge_dictionaries(dict1, dict2):
    merged_dict = {}
    for key in dict1:
        merged_dict[key] = dict1[key]
    for key in dict2:
        if key in merged_dict:
            merged_dict[key] += dict2[key]
        else:
            merged_dict[key] = dict2[key]
    return merged_dict
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 35}
result = merge_dictionaries(dict1, dict2)
print(result)
```

```python
def key_with_max_value(dictionary):
    if not dictionary:
        return None
    max_key = None
    max_value = float('-inf')
    for key, value in dictionary.items():
        if value > max_value:
            max_value = value
            max_key = key
    return max_key
my_dict = {'a': 10, 'b': 25, 'c': 15, 'd': 30}
result = key_with_max_value(my_dict)
print(f"Ключ с максимальным значением: {result}")
```

```python
def remove_keys(dictionary, keys):
    for key in keys:
        if key in dictionary:
            del dictionary[key]
    return dictionary
my_dict = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
keys = ['b', 'd', 'e']
result = remove_keys(my_dict, keys)
print(result)
```

```python
def remove_duplicate_values(dictionary):
    seen_values = set()
    result_dict = {}
    for key, value in dictionary.items():
        if value not in seen_values:
            result_dict[key] = value
            seen_values.add(value)
    return result_dict
my_dict = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}
result = remove_duplicate_values(my_dict)
print(f"Словарь без дубликатов значений: {result}")
```

```python
def set_intersection(set1, set2):
    intersection = set()
    for element in set1:
        if element in set2:
            intersection.add(element)
    return intersection
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
result = set_intersection(set_a, set_b)
print(f"Пересечение множеств: {result}")
```

```python
def set_difference(set1, set2):
    difference = set()
    for element in set1:
        if element not in set2:
            difference.add(element)
    return difference
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
result = set_difference(set_a, set_b)
print(f"Разность множеств: {result}")
```

```python
def symmetric_difference(set1, set2):
    result = set()
    for element in set1:
        if element not in set2:
            result.add(element)
    for element in set2:
        if element not in set1:
            result.add(element)
    return result
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
result = symmetric_difference(set_a, set_b)
print(f"Элементы, которые есть только в одном из множеств: {result}")
```

```python
def count_element_frequencies(input_list):
    frequency_dict = {}
    unique_elements = set(input_list)
    for element in unique_elements:
        frequency_dict[element] = input_list.count(element)
    return frequency_dict
my_list = [1, 2, 2, 3, 4, 4, 4, 5]
result = count_element_frequencies(my_list)
print(f"Частота элементов: {result}")
```

```python
def reverse_tuple(input_tuple):
    reversed_tuple = ()
    for element in reversed(input_tuple):
        reversed_tuple += (element,)
    return reversed_tuple
my_tuple = (1, 2, 3, 4, 5)
result = reverse_tuple(my_tuple)
print(f"Кортеж в обратном порядке: {result}")
```

```python
def sum_numeric_elements(input_tuple):
    sum = 0
    for element in input_tuple:
        if isinstance(element, (int, float)):
            sum += element
    return sum
my_tuple = (1, 2, 'a', 3.5, 'hello', 4)
result = sum_numeric_elements(my_tuple)
print(f"Сумма числовых элементов: {result}")
```

```python
def flatten_tuple(input_tuple):
    flat_list = []
    for element in input_tuple:
        if isinstance(element, tuple):
            flat_list.extend(flatten_tuple(element))
        else:
            flat_list.append(element)
    return tuple(flat_list)
nested_tuple = (1, (2, 3), (4, (5, 6)), 7)
result = flatten_tuple(nested_tuple)
print(f"Плоский кортеж: {result}")
```

```python
def string_to_tuple(input_string):
    string_list = input_string.split(',')
    number_tuple = tuple(float(num) if '.' in num else int(num) for num in string_list)
    return number_tuple
input_string = "1, 2.5, 3, 4.75, 5"
result = string_to_tuple(input_string)
print(f"Кортеж чисел: {result}")
```

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
def primes_in_range(start, stop):
    prime_numbers = tuple(num for num in range(start, stop) if is_prime(num))
    return prime_numbers
start = 10
stop = 30
result = primes_in_range(start, stop)
print(f"Простые числа в диапазоне [{start}, {stop}): {result}")
```
