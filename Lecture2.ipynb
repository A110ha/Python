```python
class Document:
  def __init__(self, name, pages):
    self.name=name
    self.pages=pages
  def display_info(self):print(f"Название: {self.name}, Страниц: {self.pages}")
class Recipe(Document):
  def __init__(self, name, pages, time, ingreds):
    super().__init__(name, pages)
    self.time=time
    self.ingreds=ingreds
  def set_time(self, new_time): self.time=new_time
  def display_info(self):print(f"Название: {self.name}, Страниц: {self.pages}, Время приготовления: {self.time}, Ингредиенты: {self.ingreds}")
if __name__=="__main__":
  recept=Recipe("Сливочное пюре", 2, 40, "Картофель, сливочное масло, соль")
  recept.display_info()
  recept.set_time(30)
  recept.display_info()
```

```python
class Recipe:
  def __init__(self,name,ingredients,cookingTime):
    self._name=name
    self._ingredients=ingredients
    self._cookingTime=cookingTime
  def get_name(self): return self._name
  def get_ingredients(self): return self._ingredients
  def get_cookingTime(self): return self._cookingTime
  def set_ingredients(self, new_ingredients): self._ingredients=new_ingredients
  def set_cookingTime(self, new_cookingTime): self._cookingTime=new_cookingTime
  def display_info(self): print(f"Название: {self._name}\nИнгредиенты: {self._ingredients}\nВремя приготовления: {self._cookingTime}\n")
if __name__=="__main__":
  recept=Recipe("Щи", "Капуста, картофель, морковка, лук", 90)
  recept.display_info()
  recept.set_cookingTime(100)
  recept.set_ingredients("Капуста, картофель, морковка, лук, курица")
  recept.display_info()
```

```python
class Wheel:
  def __init__(self, diameter):
    self.diameter=diameter
class Bicycle:
  def __init__(self, name):
    self.name=name
    self.wheels=[]
  def add_wheel(self, diameter):
    if len(self.wheels) >= 2:
      print("Нельзя добавить больше 2 колёс")
    else:
      new_wheel=Wheel(diameter)
      self.wheels.append(new_wheel)
  def check_wheels(self):
    if len(self.wheels) == 0:
      print(f"У велосипеда нет колёс")
    elif len(self.wheels) == 1:
      print(f"У велосипеда одно колесо")
    else:
      if(self.wheels[0]==self.wheels[1]):
        print("Диаметры колёс равны")
      else:
        print("Диаметры колёс не равны")
  def display_info(self):
    print(f"Марка велосипеда: {self.name}")
    i=0
    while i < len(self.wheels):
      print(f"Колесо {i}: Диаметр: {self.wheels[i].diameter}")
      i=i+1
if __name__=="__main__":
  bike=Bicycle("Forward")
  bike.add_wheel(40)
  bike.add_wheel(30)
  bike.check_wheels()
  bike.add_wheel(50)
  bike.display_info()
```

```python
class Coach:
  def __init__(self,name,specialization):
    self.name=name
    self.specialization=specialization
  def show(self):
    print(f"Имя: {self.name}, Специализация: {self.specialization}")
class Club:
  def __init__(self,name):
    self.name=name
    self._coaches=[]
  def add_coach(self,coach):
    if coach in self._coaches:
      print(f"Тренер уже состоит в фитнес-клубе")
    else:
      self._coaches.append(coach)
  def display_info(self):
    for i in self._coaches:
      i.show()
if __name__=="__main__":
  fitness=Club("DDX")
  coach1=Coach("Алекс","Спец1")
  coach2=Coach("Тим","Спец2")
  fitness.add_coach(coach1)
  fitness.add_coach(coach2)
  fitness.display_info()
```

```python
from abc import ABC, abstractmethod
class Photo(ABC):
  def __init__(self, name):
    self.name=name
  @abstractmethod
  def edit(self):
    pass
class Portrait(Photo):
  def edit(self):
    print(f"Редактирование портрета {self.name}: применение фильтров для лица и коррекция освещения.")
class Landscape(Photo):
  def edit(self):
    print(f"Редактирование пейзажа {self.name}: усиление цветов и настройка контраста.")
if __name__=="__main__":
  portrait=Portrait('"Девочка с персиками"')
  landscape=Landscape('"Девятый вал"')
  portrait.edit()
  landscape.edit()
```

```python
from abc import ABC, abstractmethod
class IInvoice(ABC):
  @abstractmethod
  def updateAmount(self, amount):
    pass
  @abstractmethod
  def getInvoiceDetails(self):
    pass
class Invoice(IInvoice):
  def __init__(self, invoiceNumber, amount):
    self.invoiceNumber=invoiceNumber
    self.amount=amount
  def updateAmount(self, new_amount):
    self.amount=new_amount
  def getInvoiceDetails(self):
    return f"Номер счёта: {self.invoiceNumber}, Сумма: {self.amount}"
if __name__=="__main__":
  invoice=Invoice(2.234, 5)
  print(invoice.getInvoiceDetails())
```

```python
from abc import ABC, abstractmethod
class Recipe(ABC):
  def __init__(self, name): self.name=name
  @abstractmethod
  def prepare(self):
    pass
class SimpleRecipe(Recipe):
  def prepare(self):
    print(f'Для приготовления "{self.name}" необходимо разогреть сковороду, смазать её подсолнечным маслом, разбить в сковороду 2 яйца, посолить и жарить на слабом огне 5 минут')
class ComplexRecipe(Recipe):
  def prepare(self):
    print(f'Для приготовления "{self.name}" необходимо разогреть сковороду, смазать её подсолнечным маслом, обжарить на среднем огне нарезанный бекон до золотистой корки. Затем добавить в сковороду 2 яйца, посолить и жарить на медленном огне 5 минут. Мелко нарезать зелень, посыпать на блюдо и накрыть крышкой.')
if __name__=="__main__":
  simple=SimpleRecipe("Яичница")
  hard=ComplexRecipe("Яичница с беконом")
  simple.prepare()
  hard.prepare()
```
